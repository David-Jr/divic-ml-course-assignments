### Problem 4: Data exploration

##### 1. What percentage of the loans is paid with difficulties?

```python
import numpy as np

# Computing the proportions.
loan_outcome = np.array(dataset['TARGET'])
num_paid_loans = np.count_nonzero(loan_outcome == 0)
num_unpaid_loans = np.count_nonzero(loan_outcome == 1)

# Plotting the results
plt.figure(figsize=(5, 5))
plt.pie(np.array([num_paid_loans, num_unpaid_loans]), 
        autopct='%1.1f%%', 
        startangle=140)
plt.axis('equal')
plt.title('8.1% of loans are paid with some difficulty')
plt.show()
```

### Problem 5: (Advanced task) Posting to Notebooks

##### 2. What is the relationship between the client's salary with the loan amount requested?

```python
salary = np.array(dataset['AMT_INCOME_TOTAL'])
loan_amount = np.array(dataset['AMT_CREDIT'])

plt.plot(salary, loan_amount, "o")
plt.ylabel("Loan amount") 
plt.xlabel("Salary")
plt.title("Lower salary applicants tend \nto request larger amounts ")
plt.show()
```

##### 3. What is the marital status of the applicants?

```python
# Computing the proportions.
family_status = np.array(dataset['NAME_FAMILY_STATUS'])

civil_marriage = np.count_nonzero(family_status == 'Civil marriage')
single = np.count_nonzero(family_status == 'Single / not married')
married = np.count_nonzero(family_status == 'Married')
widow = np.count_nonzero(family_status == 'Widow')

data = np.array([civil_marriage, single, married, widow])
labels = ["Civil marriage", "Single", "Married", "Widow"]
bars = plt.bar(labels, data * 100 / len(family_status), color='skyblue')


plt.bar(labels, data, color='skyblue')
plt.title('The great majority (63.8%) of applicants are married')
plt.bar_label(bars, padding=3)
plt.xlabel('Marital Status')
plt.ylabel('Number of applicants')
```

##### 4. What age group makes the most loans?

```python
applicants_age = np.array(dataset['DAYS_BIRTH']) # Age in days, loan date - birthdate
applicants_age = applicants_age / -365

plt.violinplot(applicants_age)
plt.ylabel("Applicant's age in years")
plt.title("35 to 45 year olds are a \nlarger part of the applicants ")
plt.show()
```

##### 5. What other are some factors that contribute to the loan amount?

```python
import seaborn as sns

selected_variables: DataFrame = dataset[['TARGET', 'CNT_CHILDREN', 'AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'DAYS_REGISTRATION', 'OWN_CAR_AGE', 'CNT_FAM_MEMBERS', 'HOUR_APPR_PROCESS_START']].copy()

selected_variables['DAYS_BIRTH'] *= -1
selected_variables['DAYS_EMPLOYED'] *= -1
selected_variables['DAYS_REGISTRATION'] *= -1

sns.heatmap(selected_variables.corr(), cmap='coolwarm', annot=True, annot_kws={"fontsize":5})
plt.title("Client's income, number of family members, age and the their \ntime employed are highly correlated with the loan amount\n")
plt.show()
```
