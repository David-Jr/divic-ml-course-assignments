### Creating dummy data

#### Problem 1: Creating random numbers

```python
import numpy as np

np.random.seed(0)

mean = (-3, 0)
cov = np.array([[1.0, 0.8], [0.8, 1.0]])
size = 500
random_numbers = np.random.multivariate_normal(mean, cov, size=size)

# print(random_numbers)
```

#### Problem 2: Visualization by scatter plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Since our random_numbers have a (500, 2), we can use the first 
# column for X values and the second for Y values of our scatter plot.
x = random_numbers[:, 0]
y = random_numbers[:, 1]

plt.scatter(x, y, alpha=0.5)
plt.title('Scatter Plot of Dummy Data')
plt.xlabel('X1')
plt.ylabel('X2')
```
